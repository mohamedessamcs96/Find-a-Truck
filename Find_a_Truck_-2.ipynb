## Draw Rectangle

```python
# x,y
top_left=(100,50)
bottom_right=(450,350)
truck=cv2.rectangle(rgb_img,top_left,bottom_right,(0,0,255),5)
```

```python
plt.imshow(truck)
```

```python
img=cv2.imread('truck.jpg')
template=img[80:350,120:420]
plt.imshow(template)
```

```python
w,h=template.shape[:2]
```

```python
result=cv2.matchTemplate(img,template,cv2.TM_CCOEFF_NORMED)
result.shape
```

```python
plt.imshow(result,cmap='gray')
```

```python
min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(result)
```

```python
top_left=max_loc
bottom_right=(top_left[0]+w,top_left[1]+h)
cv2.rectangle(img,top_left,bottom_right,(255,0,0),5)
```

```python
plt.imshow(img)
```
